```python
# Code pour notebooks/01_data_exploration.ipynb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger les données nettoyées
df = pd.read_csv("/Users/corsair/Documents/Projet_AirBnB_Prediction/airbnb-price-prediction/data/processed/paris_listings_cleaned.csv")
df
```

```python
# Corrélation entre prix et minimum_nights
plt.figure(figsize=(8, 6))
sns.scatterplot(x='minimum_nights', y='price', data=df)
plt.title('Prix vs Nombre minimum de nuits')
```

```python
# Prix par quartier
plt.figure(figsize=(14, 8))
sns.boxplot(x='neighbourhood_cleansed', y='price', data=df)
plt.title('Prix par quartier')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
# Statistiques descriptives
print("\nStatistiques descriptives:")
df.describe()
```

```python

# Aperçu des données
print("Aperçu des données:")
print(df.head())

# Informations sur les colonnes
print("\nInformations sur les colonnes:")
print(df.info())

# Statistiques descriptives
print("\nStatistiques descriptives:")
df.describe()

# Valeurs manquantes
print("\nValeurs manquantes par colonne:")
print(df.isnull().sum())

# Visualisation de la distribution des prix
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], bins=50)
plt.title('Distribution des prix')
plt.xlabel('Prix')
plt.ylabel('Fréquence')
plt.show()

# Prix par quartier
plt.figure(figsize=(14, 8))
sns.boxplot(x='neighbourhood_cleansed', y='price', data=df)
plt.title('Prix par quartier')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Prix par type de logement
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Prix par type de logement')
plt.show()

# Corrélation entre prix et minimum_nights
plt.figure(figsize=(8, 6))
sns.scatterplot(x='minimum_nights', y='price', data=df)
plt.title('Prix vs Nombre minimum de nuits')
plt.show()

# Distribution du nombre minimum de nuits
plt.figure(figsize=(10, 6))
sns.histplot(df['minimum_nights'], bins=30)
plt.title('Distribution du nombre minimum de nuits')
plt.show()
```
